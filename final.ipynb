# Using Elbow algorithm to get better K value

```python
   # Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans

# to store WCSS
wcss = []

# for loop 
for i in range(1, 11): # k-mean cluster model for different k values
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(train_input_SMOTE,train_output_SMOTE)
    
    # inertia method returns wcss for that model
    wcss.append(kmeans.inertia_)
    
# importing the matplotlib module
import matplotlib.pyplot as plt

# figure size
plt.figure(figsize=(10,5))
sns.lineplot(range(1, 11), wcss,marker='o',color='green')

# labeling
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Visualization of clustered data
# k-mean algorithm
kmeans = KMeans(n_clusters = 2, random_state = 42)
y_kmeans = kmeans.fit_predict(train_input_SMOTE,train_output_SMOTE)
X = np.array(train_input_SMOTE,train_output_SMOTE)

# size of image
plt.figure(figsize=(10,7))

# visualizing the clusters
sns.scatterplot(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], color = 'yellow', label = 'Cluster 1',s=50)
sns.scatterplot(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], color = 'blue', label = 'Cluster 2',s=50)

sns.scatterplot(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color = 'red', 
                label = 'Centroids',s=300,marker=',')

# labeling
# plt.grid(False)
plt.title('Two clusters')
plt.xlabel('train_output_SMOTE')
plt.ylabel('values')
plt.legend()
plt.show()
```
